```python
# print 3 random attractions in the dataset

import re

def extract_attractions(text):
    
    # convert to string if it is not a string
    text = str(text)
    
    # Entferne die erste Zeile bis zum ersten <br />
    text = re.sub(r'^.*?<br />', '', text, count=1).strip()
    
    # Entferne alle Vorkommen von <p> und </p>
    text = text.replace('</p>', '').replace('<p>', '')
    
    # Splitte den Text basierend auf <br /> als Delimiter
    parts = text.split('<br />')
    
    attractions = []
    preferred_airport = None
    
    attraction_pattern = re.compile(r"(.*?) - (\d+\.\d+) km / (\d+\.\d+) mi")
    preferred_airport_pattern = re.compile(r".*? airport is (.*?) - (\d+\.\d+) km / (\d+\.\d+) mi")
    
    for part in parts:
        part = part.strip()
        airport_match = preferred_airport_pattern.search(part)
        
        if airport_match:
            airport_name, km_distance, mi_distance = airport_match.groups()
            preferred_airport = {
                "airportName": airport_name.strip(),
                "distanceKm": float(km_distance),
                "distanceMi": float(mi_distance)
            }
            continue
        
        match = attraction_pattern.match(part)
        if match:
            attraction_name, km_distance, mi_distance = match.groups()
            attractions.append({
                "attractionName": attraction_name.strip(),
                "distanceKm": float(km_distance),
                "distanceMi": float(mi_distance)
            })
    
    return {
        "attractions": attractions,
        "preferredAirport": preferred_airport
    }

attr = df.sample(3)['Attractions']

for a in attr:
    print(a)
    print("### Extracted attractions:")
    print(extract_attractions(a))
    print ("---")
```

```python
# try to extract all attractions from the dataset

all_attractions = df['Attractions'].apply(extract_attractions)
```

```python
# print 5 random attractions in the dataset

rnd_attractions = all_attractions.sample(5)

for i, attraction in enumerate(rnd_attractions):
    print(f"Attractions {i}:")
    print(attraction)
    print("---")
```

The hotels.csv has following format:

| Attribute        | Value Range                                  |
|-----------------|---------------------------------------------|
| countyCode     | Two-letter country code (ISO 3166-1 alpha-2), e.g., "US", "DE", "FR" |
| countyName     | Name of the country, e.g., "Germany", "United States" |
| cityCode        | Unique city code, e.g., "NYC", "BER" |
| cityName        | Name of the city, e.g., "Berlin", "New York" |
| HotelCode       | Unique hotel code, numeric, e.g., "1003300" |
| HotelName       | Name of the hotel, e.g., "Hotel Parlamenti" |
| HotelRating     | 'FourStar' 'ThreeStar' 'All' 'FiveStar' 'TwoStar' 'OneStar' |
| Address         | Full address of the hotel |
| Attractions     | Attractions are separated with <br />. Each attraction has attributes "attractionName", "distanceKm", "distanceMi". Additionally, the preferred airport can be extracted from the last line with attributes "airportName", "distanceKm", "distanceMi" |
| Description     | Description of the hotel, including amenities and features |
| FaxNumber       | Fax number, e.g., "00355 (0)42273689" |
| HotelFacilities | List of hotel facilities, e.g., "Pool, WiFi, Spa" (no specific format that can be extracted) |
| Map            | coordinates of the hotel, e.g., "41.32213|19.81665" |
| PhoneNumber     | Phone number, e.g., "00355 4226 5009" |
| PinCode        | Postal code of the hotel, e.g., "1000" |
| HotelWebsiteUrl | URL of the official hotel website, e.g., "https://www.booking.com/hotel/al/de-paris.html" |
